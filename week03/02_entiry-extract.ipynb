### read stored data from DataPlatform

```python
from dsmlibrary.datanode import DataNode
import json
from tqdm.auto import tqdm
import pandas as pd
```

```python
dir_rawData_id = 87
dir_processData_id = 89
```

```python
datanode = DataNode()
```

```python
meta, fp = datanode.get_file(datanode.get_file_id(directory_id=dir_rawData_id, name='thaigem.json'))
meta
```

```python
data = json.load(fp)
data
```

```python
df = pd.json_normalize(data)
```

### write DataNode to DataPlatform

```python
datanode.write( df=df, 
                directory=dir_processData_id, 
                name="week03_thaigem", 
                profiling=True, 
                lineage=[datanode.get_file_id(directory_id=dir_rawData_id, name='bkkbiz.json')],
                description="3:  , ,   8/11/2022"
              )
```
